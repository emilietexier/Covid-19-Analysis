## Covid-19 Analysis

#### Importation des librairies

```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

#### Pre-processing du dataset

```python
f1 = open(r'C:\Users\emili\Desktop\covid-19_data\data\worldwide-aggregated_csv.csv')
f2 = open(r'C:\Users\emili\Desktop\covid-19_data\data\countries-aggregated_csv.csv')

data_world = csv.reader(f1)
data_coutries = csv.reader(f2)

#185 pays
#date, confirmed, recovered, deaths, increase rate
#date, country, confirmed, recovered, deaths

confirmed_world = []
recovered_world = []
deaths_world = []
increase_rate_world = []

for row in data_world:
    if row[0] != 'Date':
        confirmed_world.append(row[1])
        recovered_world.append(row[2])
        deaths_world.append(row[3])
        increase_rate_world.append(row[4])
    
confirmed_world = [float(i) for i in confirmed_world]
recovered_world = [float(i) for i in recovered_world]
deaths_world = [float(i) for i in deaths_world]
increase_rate_world[0] = 0
n = len(increase_rate_world) - 1
for i in range(n):
    increase_rate_world[i + 1] = float(increase_rate_world[i + 1])
```

#### Analyse au niveau mondial

```python
days = range(83)

plt.plot(days, confirmed_world)
plt.xlabel('Jours')
plt.ylabel('Nombre de cas')
plt.title('Evolution du nombre de cas dans le monde')
plt.show()

plt.plot(days, recovered_world)
plt.xlabel('Jours')
plt.ylabel('Nombre de guerisons')
plt.title('Evolution du nombre de guerisons dans le monde')
plt.show()

plt.plot(days, deaths_world)
plt.xlabel('Jours')
plt.ylabel('Nombre de morts')
plt.title('Evolution du nombre de morts dans le monde')
plt.show()
```
